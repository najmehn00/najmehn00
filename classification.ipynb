```python
#library
import numpy as np
import pandas as pd

#dataset
data = pd.read_csv (r'/Users/najmeh/Desktop/master/Ml/projet/secound proj/classification/Data_Normal.txt')
data.to_csv (r'/Users/najmeh/Desktop/master/Ml/projet/secound proj/classification/Data_Normal.csv', index=None)
data.shape
print(data)


#a function for predict
def predict(df,y,thresh_hold):
    y_pred=[]
    for label in df[y]:
        if label<thresh_hold:
            y_pred.append(0)
        else:
            y_pred.append(1)
    return y_pred


# confusion matrix
def cal_vals(df,y,y_pred):
    tp=0
    tn=0
    fn=0
    fp=0
    for val1,val2 in enumerate(df['y']):
        if(df.y_pred[val1]==1) and df.y[val1]==1:
            tp=tp+1
        if(df.y_pred[val1]==0) and df.y[val1]==0:
            tn=tn+1
        if(df.y_pred[val1]==0) and df.y[val1]==1:
            fn=fn+1
        if(df.y_pred[val1]==1) and df.y[val1]==0:
            fp=fp+1
    return {'tn':tn,'tp':tp,'fn':fn,'fp':fp}



thresh_hold=0.5
data['y_pred']=predict(data,'Male',thresh_hold)
confusion_matrix=cal_vals(data,'y','y_pred')
    


# confusion matrix values 
print("the confusion matrix is: ",confusion_matrix)

```
