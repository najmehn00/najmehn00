```python
#library
import numpy as np
import pandas as pd


#load dataset
filename = 'Data_Normal.txt'
data = np.loadtxt(filename, delimiter=',', skiprows=1, dtype=str)


def mapping(data,feature):
    featureMap=dict()
    count=0
    for i in sorted(data[feature].unique(),reverse=True):
        featureMap[i]=count
        count=count+1
    data[feature]=data[feature].map(featureMap)
    return data

#normalize data
arr=np.array([89,78,14,1619,20])
result=np.linalg.norm(arr)
result=float(result)
newoutput=arr/result
print(newoutput)







```
